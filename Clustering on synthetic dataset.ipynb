```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, MeanShift, estimate_bandwidth
from sklearn.mixture import GaussianMixture
import scipy.cluster.hierarchy as sch

# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

```

```python
# K-Means Clustering
kmeans = KMeans(n_clusters=4)
y_kmeans = kmeans.fit_predict(X)
plt.figure()
plt.title('K-Means Clustering')
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', s=200, alpha=0.75)
plt.show()

```

```python
# Hierarchical Clustering
linked = sch.linkage(X, method='ward')
plt.figure(figsize=(10, 7))
plt.title('Hierarchical Clustering Dendrogram')
sch.dendrogram(linked, truncate_mode='level', p=5)
plt.show()
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
y_hc = hc.fit_predict(X)
plt.figure()
plt.title('Hierarchical Clustering')
plt.scatter(X[:, 0], X[:, 1], c=y_hc, s=50, cmap='viridis')
plt.show()

```

```python
# DBSCAN Clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(X)
plt.figure()
plt.title('DBSCAN Clustering')
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, s=50, cmap='viridis')
plt.show()


```

```python
# Mean Shift Clustering
bandwidth = estimate_bandwidth(X, quantile=0.2, n_samples=300)
ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
y_ms = ms.fit_predict(X)
plt.figure()
plt.title('Mean Shift Clustering')
plt.scatter(X[:, 0], X[:, 1], c=y_ms, s=50, cmap='viridis')
plt.scatter(ms.cluster_centers_[:, 0], ms.cluster_centers_[:, 1], c='red', s=200, alpha=0.75)
plt.show()

```

```python
#import warnings
#warnings.filterwarnings('ignore', category=UserWarning, module='threadpoolctl')


# Gaussian Mixture Model Clustering
gmm = GaussianMixture(n_components=4)
y_gmm = gmm.fit_predict(X)
plt.figure()
plt.title('Gaussian Mixture Model Clustering')
plt.scatter(X[:, 0], X[:, 1], c=y_gmm, s=50, cmap='viridis')
plt.show()
```
